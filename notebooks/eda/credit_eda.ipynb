# Credit dataset: initial EDA

This notebook examines the Default of Credit Card Clients dataset from the UCI Machine Learning Repository.

According to the dataset documentation, the dataset contains 23 input features and one target variable, `Y`, resulting in 24 columns in total. All variables are represented numerically, although some of them correspond to categorical attributes encoded with integer values. The dataset does not contain missing values.

```python
import pandas as pd

from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
default_of_credit_card_clients = fetch_ucirepo(id=350) 
  
# data (as pandas dataframes) 
X = default_of_credit_card_clients.data.features 
y = default_of_credit_card_clients.data.targets 
  
# metadata 
#print(default_of_credit_card_clients.metadata) 
  
# variable information 
print(default_of_credit_card_clients.variables) 

df = pd.concat([X, y], axis=1)
```

## Dataset overview

The dataset contains **30,000 rows** and **24 columns**.

The first 23 columns are input features, while the last column, `Y`, is the target variable.

```python
df.describe()
```

## Numerical columns

The numerical summary shows that the main variables have plausible ranges.

A few initial observation(s):
- `X5: Age` ranges from **21** to **79**, which is reasonable for this dataset.

At this stage, there are no obvious invalid values in the numerical columns, but the distributions of some variables may still require closer inspection through visualization.

```python
for n in ['X2', 'X3', 'X4', 'Y']:
    print("List of attributes of category", n, ":", df[n].unique())
    print(df[n].value_counts())
```

## Categories

The categorical variables inspected here are `X2`, `X3`, `X4`, and `Y`.

The target variable `Y` contains only the values **0** and **1**, which is consistent with the dataset documentation (`0 = no default`, `1 = default`). However, the distribution is clearly imbalanced, since non-default cases are much more frequent than default cases.

`X2` also contains only two distinct values, **1** and **2**, corresponding to the two documented gender categories. Although this variable is valid as documented, it is also imbalanced, with one category appearing more frequently than the other.

For `X3` (education) and `X4` (marital status), the observed categories do not fully match the original variable description. According to the documentation, `X3` should contain values **1-4**, but the dataset also includes **0, 5, and 6**. Similarly, `X4` is documented with values **1-3**, yet the dataset also contains the value **0**. These additional codes are not described in the original variable information and should therefore be noted as undocumented category values that may require special handling in later preprocessing.

## Sensitive attributes

The main sensitive attribute in this dataset is `X2:Gender`.

These attributes are important because they are related to fairness and representation. If some groups are much smaller than others, a synthetic data generator may reproduce or even amplify these imbalances.

Gender is imbalanced, with more **Female** than **Male** records.

These patterns should be considered later when evaluating whether the synthetic data preserves minority groups appropriately.

```python
pay_cols = ["X18", "X19", "X20", "X21", "X22", "X23"]

df[pay_cols].hist(figsize=(12, 8), bins=50)
plt.suptitle("Histograms of previous payment amounts")
plt.tight_layout()
plt.show()
```

## Payment amounts

The previous payment amount variables are also highly right-skewed, with many observations concentrated near lower values and a smaller number of large payments forming long tails. Their similar shapes reflect the fact that they capture the same type of financial behavior across multiple months.

## Repayment Status

The repayment status variables `X6`-`X11` show very similar category distributions across the six months, which is expected because they represent the same type of monthly repayment behavior. Most observations are concentrated around a small number of status codes, while larger delay values occur much less frequently. This suggests that severe repayment delays are relatively rare compared to timely repayment or only minor delays.

```python
import seaborn as s
s.heatmap(df.corr(), cmap="coolwarm")
```

## Correlation

The correlation heatmap shows several clear patterns in the dataset. The repayment status variables X6-X11 are positively correlated with each other, which is expected since they describe the same type of payment behavior across consecutive months. This suggests that clients who experienced repayment difficulties in one month were more likely to show similar behavior in other months as well.

An even stronger correlation pattern appears among the bill statement amount variables X12-X17. These variables are highly positively correlated, indicating that clients with high bill amounts in one month tended to have high bill amounts in the surrounding months too. This is reasonable, since monthly bill amounts usually follow a relatively stable financial pattern over time.

In addition, X1 (LIMIT_BAL) shows a slight negative correlation with the repayment status variables X6-X11. This suggests that clients with higher credit limits tended to have somewhat lower repayment-status values, meaning fewer or less severe delays on average. However, the relationship is weak, so it should be interpreted cautiously.

```python
rare_cols = ["X2", "X3", "X4"]

for col in rare_cols:
    print(df[col].value_counts().sort_index(), df[col].value_counts(normalize=True).sort_index() * 100,'\n')
```

## Rare categories

For X3 (Education: 1 = graduate school, 2 = university, 3 = high school, 4 = others), the distribution is unbalanced. The main categories are university (2, 46.77%), graduate school (1, 35.28%), and high school (3, 16.39%), which together make up almost the entire dataset. The documented category 4 (“others”) appears only in 0.41% of observations, so it can be considered rare. In addition, the dataset contains the undocumented categories 0 (0.05%), 5 (0.93%), and 6 (0.17%), which are also very rare and are not explained in the original variable description.

For X4 (Marital status: 1 = married, 2 = single, 3 = others), the dataset is dominated by single (2, 53.21%) and married (1, 45.53%) clients. The documented category 3 (“others”) appears in only 1.08% of observations, making it relatively rare. The undocumented value 0 is even less frequent, appearing in just 0.18% of records.

Overall, the documented “others” categories are already uncommon, and the additional undocumented values are even rarer. These rare and unexplained categories should be noted carefully, since they may need to be merged into an “other” or “unknown” group during preprocessing.

## Initial preprocessing considerations

Based on this EDA, the main preprocessing issues are:
- deciding how to treat duplicate rows
- paying attention to strong class imbalance in `Y`, `X2`, `X3` and `X4`
- deciding how to handle undocumented category values in `X3` and `X4`
- encoding categorical variables appropriately for baseline method(s)
- monitoring highly skewed numerical variables such as `X1` and the monthly bill and payment amount variables (`X12`-`X23`)
- considering the strong correlation among monthly repayments, bill statements (`X6`-`X23`) during later modeling
